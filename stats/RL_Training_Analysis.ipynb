# RL Training Analysis

本 Notebook 用來讀取訓練期間儲存的各項指標 (.npy 檔案)，並製作圖表來分析 agent 的表現。這些指標包括：

- **Episode Rewards**: 每個 episode 的累積 reward
- **Cash History**: 每個 episode 結束時的現金餘額
- **Position History**: 每個 episode 結束時的持倉量
- **Loss History**: 每個 episode 的平均 loss
- **Step Reward 分布**: 每個 step 的 reward 分布

我們將根據這些圖表來評估 agent 是否穩定獲利，以及哪些部分需要進一步改進。

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# 載入 npy 檔案 (請確認這些檔案與 Notebook 在同一個資料夾下)
episode_rewards = np.load('./episode_rewards.npy')
cash_history = np.load('./cash_history.npy')
position_history = np.load('./position_history.npy')
actor_history = np.load('./actor_loss_history.npy')
step_rewards = np.load('./step_rewards.npy')
action0 = np.load("./action0_history.npy")
cumulative_return = np.load('./cumulative_return.npy')

temp = 10.0
```

## Episode-level Analysis

```python
# 繪製 Episode Reward 曲線
plt.figure(figsize=(10,5))
plt.plot(episode_rewards, marker='o')
plt.title('Episode Rewards')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.show()

# 繪製 Cash History 曲線
plt.figure(figsize=(10,5))
plt.plot(cash_history, marker='o', color='green')
plt.title('Cash History')
plt.xlabel('Episode')
plt.ylabel('Cash')
plt.show()

# 繪製 Position History 曲線
plt.figure(figsize=(10,5))
plt.plot(position_history, marker='o', color='orange')
plt.title('Position History')
plt.xlabel('Episode')
plt.ylabel('Position (Shares)')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(actor_history, marker='o', color='green')
plt.title('actor Loss History')
plt.xlabel('Episode')
plt.ylabel('Loss')
plt.show()
```

## Step-level Analysis

```python


# 繪製 Step Rewards 的直方圖
plt.figure(figsize=(10,5))
plt.hist(step_rewards, bins=200, color='purple', alpha=0.7)
plt.title('Histogram of Step Rewards')
plt.xlabel('Step Reward')
plt.ylabel('log')
plt.show()

action_dim = 11
plt.hist(action0, bins=range(action_dim+1), align='left', rwidth=0.8)
plt.xlabel("Action index")
plt.ylabel("Frequency")
plt.title(f"Action[0] Distribution under Temperature={temp}")
plt.show()

plt.figure(figsize=(10,5))
plt.plot(cumulative_return, marker='o', color='orange')
plt.title('cumulative return')
plt.xlabel('Episode')
plt.ylabel('$')
plt.show()
```
